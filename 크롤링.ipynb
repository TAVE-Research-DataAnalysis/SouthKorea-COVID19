```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re
import time
```

```python
def insta_searching(word):
    url ="https://www.instagram.com/explore/tags/" + str(word)
    return url
```

```python
def select_first(driver):
    first = driver.find_elements_by_css_selector("div.EZdmt")[0]
    first.click()
    time.sleep(3)
```

```python
def get_content(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    try:
        content = soup.select('div.C4VMK > span')[0].text
    except:
        content = ''
    tags = re.findall(r'#[^\s#, \\]+', content)
    date = soup.select('time._1o9PC.Nzb55')[0]['datetime'][:10]
    try:
        like = soup.select('a.zV_Nj')[0].text[4:-1]
    except:
        like = 0
    try:
        place = soup.select('div.M30cS')[0].text
    except:
        place = ''
    data = [content, date, like, place, tags]
    return data
```

```python
def move_next(driver):
    right = driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow')
    right.click()
    time.sleep(10)
```

```python
driver = webdriver.Chrome('C:/Users/keogu/Desktop/TAVE_RESEARCH/chromedriver_win32/chromedriver.exe')

driver.get('https://www.instagram.com')
time.sleep(10)

email = '계정 아이디(이메일) 입력'
input_id = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[0]
input_id.clear()
input_id.send_keys(email)

password = '비밀번호 입력'
input_pw = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[1]
input_pw.clear()
input_pw.send_keys(password)
input_pw.submit()

time.sleep(5)

word = input("검색어를 입력하세요:")
word = str(word)
url = insta_searching(word)

driver.get(url)
time.sleep(10)

select_first(driver)

result = []
target = 10000
for i in range(target):
    try:
        data = get_content(driver)
        result.append(data)
        move_next(driver)
    except:
        time.sleep(2)
        move_next(driver)
        
print(result[:2])
```

```python
import pandas as pd
df_result = pd.DataFrame(result)
```

```python
df_result.rename(columns = {0:'content', 1:'date', 2:'like', 3:'place', 4:'tags'},inplace=True)
```
